```python
import pandas as pd

cellulosome = pd.read_excel("prot_trans_const_file.xlsx", sheet_name = "Cellulosome Requirements")
metabolic = pd.read_excel("prot_trans_const_file.xlsx", sheet_name = "Metabolic Requirements")
```

```python
with open('prot_exp_flux_cell.txt', 'w') as file:
    line = f"/\n"
    file.write(line)
    for index, row in cellulosome.iterrows():
        line = f"'PROSYN-{row['Cellulosome Protein']}'\t{row['Flux (mmol/(gDW hr)']}\n"
        file.write(line)
    file.write("/")
```

```python
with open('prot_exp_flux_met.txt', 'w') as file:
    line = f"/\n"
    file.write(line)
    for index, row in metabolic.iterrows():
        line = f"'PROSYN-{row['Metabolic Protein']}'\t{row['Flux (mmol/(gDW hr)']}\n"
        file.write(line)
    file.write("/")
```
